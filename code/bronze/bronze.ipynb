# Bronze fiels ETL

Useing pyspark

## Assign the names and paths for the files

```python
files_config =[
    {
        "table_name":"crm_cust_info",
        "path":"/Volumes/workspace/bronze/source_system/source_crm/cust_info.csv"
    },
    {
        "table_name":"prd_info",
        "path":"/Volumes/workspace/bronze/source_system/source_crm/prd_info.csv",
    },
    {
        "table_name":"sales_details",
        "path":"/Volumes/workspace/bronze/source_system/source_crm/sales_details.csv",
    },
    {
        "table_name":"CUST_AZ12",
        "path":"/Volumes/workspace/bronze/source_system/source_erp/CUST_AZ12.csv",
    },
    {
        "table_name":"LOC_A101",
        "path":"/Volumes/workspace/bronze/source_system/source_erp/LOC_A101.csv",
    },
    {
        "table_name":"PX_CAT_G1V2",
        "path":"/Volumes/workspace/bronze/source_system/source_erp/PX_CAT_G1V2.csv",
    },
]
```

## Read and write the files

```python
dfs = {}

for file in files_config:
    table = file["table_name"]
    path = file["path"]
    # Read CSV into DataFrame
    dfs[table] = spark.read.option("header", "true").option("inferSchema", "true").csv(path)
    
    print(f"{table} is loaded")
    # Save DataFrame as Delta table
    dfs[table].write.format("delta").mode("overwrite").saveAsTable(f"workspace.bronze.{table}")
    print(f"{table} is saved")
```
