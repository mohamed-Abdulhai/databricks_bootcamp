Reading the file

```python
df = spark.read.table("bikes.bronze.erp_px_cat_g1v2")
```

Importing the libraries

```python
from pyspark.sql.functions import when, col, StringType, trim
```

triming the columns

```python
for field in df.schema:
  if field.dataType == StringType():
    df = df.withColumn(field.name, trim(col(field.name)))
```

Clean the values and rename columns names

```python
# clean fun
df_clean = (
    df

    # Rename columns
    .withColumnRenamed("ID", "product_category_key")
    .withColumnRenamed("CAT", "category_name")
    .withColumnRenamed("SUBCAT", "subcategory_name")
    .withColumnRenamed("MAINTENANCE", "maintenance_required")

    .withColumn(
    "maintenance_required",
    when(col("maintenance_required") == "Yes", True)
    .otherwise(False)
    )
)
df = df_clean
```

Save the table

```python
df.write.mode("overwrite").saveAsTable("bikes.silver.erp_category")
```
