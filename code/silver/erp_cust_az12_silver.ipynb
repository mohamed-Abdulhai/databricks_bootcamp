Read the file

```python
df = spark.read.table("bikes.bronze.erp_cust_az12")
```

Rename the columns

```python
# clean fun
df_clean = (
    df
    .withColumnRenamed("CID", "customer_id")
    .withColumnRenamed("BDATE", "date_of_birth")
    .withColumnRenamed("GEN", "gender")
)
df = df_clean
```

Save the table

```python
df.write.mode("overwrite").saveAsTable("bikes.silver.erp_customer")
```
