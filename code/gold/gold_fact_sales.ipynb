# **fact_sales Table**

Start by create the table and adding the comments

```python
%sql CREATE TABLE IF NOT EXISTS bikes.gold.fact_sales (
  order_number STRING COMMENT 'Unique identifier for the order',
  customer_id STRING COMMENT 'Unique identifier for the customer',
  product_id STRING COMMENT 'Unique identifier for the product',
  order_date DATE COMMENT 'Date of the order',
  ship_date DATE COMMENT 'Date of the shipping',
  due_date DATE COMMENT 'Date of the due',
  quantity INT COMMENT 'Quantity of the product',
  unit_price DECIMAL(10,2) COMMENT 'Unit price of the product',
  total_sales DECIMAL(10,2) COMMENT 'Total sales of the order'
  )
  COMMENT 'Table containing the sales data';
```

Insert the date 

```python
%sql
INSERT OVERWRITE bikes.gold.fact_sales
SELECT
  sls_ord_num,
  sls_cust_id,
  sls_prd_key,
  try_to_date(sls_order_dt, 'yyyyMMdd'),
  try_to_date(sls_ship_dt, 'yyyyMMdd'),
  try_to_date(sls_due_dt, 'yyyyMMdd'),
  sls_quantity,
  sls_price,
  sls_sales
FROM bikes.silver.crm_sales_details;
```

Add the tags

```python
%sql
ALTER TABLE bikes.gold.fact_sales
SET TAGS (
  'layer' = 'gold',
  'table_type' = 'fact',
  'domain' = 'sales',
  'pii' = 'false'
);
```
